## Phase velocity and the group velocity polar plots

> Plot the phase velocity and the group velocity polar plots for $C_S= 0.25 C_A$ and $C_S= 4 C_A$.


```julia
import Pkg;
Pkg.add("Optim");
```

```julia
include("main.jl")
using AlgebraOfGraphics
using CairoMakie
using beforerr
```

```julia
maximum_vgs_angle
```

```julia
dfs = [0.8, 1, 1.25] .|> VpVg_fastandslow_df
df = vcat(dfs...);
```

```julia
ca = 1
cs = 0.8
θmax = maximum_vgs_angle(cs, ca)
Vgs_para, Vgs_perp = vgs(θmax, cs, ca)
```

```julia
axis = (width=225, height=225)

base_map = data(df) * mapping(
    :v_parp => L"V_{\parallel}",
    :v_perp => L"V_{\perp}",
) * visual(Scatter, markersize=3);
```

```julia
df = VpVg_fastandslow_df(cs)

base_map = data(df) * mapping(
    :v_parp => L"V_{\parallel}",
    :v_perp => L"V_{\perp}",
) * visual(Scatter, markersize=3);

plt = base_map * mapping(color=:wave, marker=:type, markersize=:type, col=:cs => nonnumeric)
fig = draw(plt, axis=axis)
lines!(
    [0, Vgs_para], [0, Vgs_perp],
    color=:black,
    linestyle=:dash,
)
# easy_save("mhd_waves_max")
fig
```
