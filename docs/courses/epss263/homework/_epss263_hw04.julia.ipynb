## Phase velocity and the group velocity polar plots

> Plot the phase velocity and the group velocity polar plots for $C_S= 0.25 C_A$ and $C_S= 4 C_A$.


```julia
include("main.jl")
using AlgebraOfGraphics
using CairoMakie
using beforerr
```

```julia
dfs = [0.25, 1, 4] .|> VpVg_fastandslow_df
df = vcat(dfs...);
```

```julia
axis = (width=225, height=225)
xy_map = mapping(
    :v_parp => L"V_{\parallel}",
    :v_perp => L"V_{\perp}",
) * visual(Scatter, markersize=3);
base_map = data(df) * xy_map;
```

```julia
plt = base_map * mapping(color=:wave, marker=:wave, row=:cs => nonnumeric, col=:type)
draw(plt, axis=axis)
easy_save("mhd_waves_1")
```

```julia
plt = base_map * mapping(col=:wave, row=:cs => nonnumeric, color=:type)
fg = draw(plt, axis=axis)
easy_save("mhd_waves_2")
```

```julia
plt = base_map * mapping(color=:wave, marker=:type, markersize=:type, col=:cs => nonnumeric)
fg = draw(plt, axis=axis)
easy_save("mhd_waves_4")
```

```julia
plt = base_map * mapping(row=:wave, color=:cs => nonnumeric, col=:type)
fg = draw(plt, axis=axis)
easy_save("mhd_waves_3")
```

```julia
for sym in [:SlowWave, :FastWave]
    @eval function $sym(cs, ca)
        cms = sqrt(cs^2 + ca^2) #: magnetosonic speed
        return $sym(cs, ca, cms)
    end

    @eval $sym(cs) = $sym(cs, 1)
end
```

```julia
function plot_maximum_vgs_angle(cs)
    df = VpVg_fastandslow_df(cs)

    wave = SlowWave(cs)

    θmax = maximum_vgs_angle(cs)
    phi = vgs_angle(θmax, cs)

    @show θmax phi

    Vps_para, Vps_perp = phase_velocity(wave, θmax)
    Vgs_para, Vgs_perp = vgs(θmax, cs)

    subset!(df, :type => ByRow(==("group velocity")))
    base_map = data(df) *xy_map;
    plt = base_map * mapping(color=:wave, col=:cs => nonnumeric)
    fig = draw(plt, axis=axis)

    lines!([0, Vps_para], [0,Vps_perp], color=:black)
    lines!(
        [0, Vgs_para], [0, Vgs_perp],
        color=:black,
        linestyle=:dash,
    )
    return fig
end
```

```julia
plot_maximum_vgs_angle(0.8)
easy_save("vgs_angle_1")
```

```julia
plot_maximum_vgs_angle(1.25)
easy_save("vgs_angle_2")
```
